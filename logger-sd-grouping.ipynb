```python
import pandas as pd

# Assuming df is already defined
df_time = df[
    ["No", "Title", "Date Created", "time measure start", "time measure end"]
].copy()

# Create new columns
df_time.loc[:, "date created start"] = df_time["Date Created"]
df_time.loc[:, "date created end"] = df_time["Date Created"].shift(-1)

# Handle the last row separately
last_row_index = df_time.index[-1]
df_time.at[last_row_index, "date created end"] = (
    df_time.at[last_row_index, "Date Created"].split("T")[0]
    + "T"
    + str(df_time.at[last_row_index, "time measure end"])
    + "+07:00"
)

# Convert 'date created start' and 'date created end' to datetime
df_time["date created start"] = pd.to_datetime(df_time["date created start"])
df_time["date created end"] = pd.to_datetime(df_time["date created end"])

# make column `is_time_inside` to check if the time is inside the range
df_time.loc[:, "is_time_inside"] = df_time.apply(
    lambda row: row["time measure start"] >= row["date created start"].time()
    and row["time measure end"] <= row["date created end"].time(),
    axis=1,
)

# Set display options to show all rows and columns
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

# Show all the rows
df_time
```

# prepare data logger

```python
import os
import pandas as pd

# Folder containing the .TXT files
log_folder = "logger-sd-lapangan"

# Initialize an empty list to store the log data
log_data = []

# Read the .TXT files and extract the values
index_log_file = -1
for log_file in os.listdir(log_folder):
    index_log_file += 1
    # print(f"Reading file: {log_file}")
    if log_file.endswith(".TXT"):
        with open(os.path.join(log_folder, log_file), "r") as file:
            for line in file:
                try:
                    if "=" in line:
                        continue
                    parts = line.strip().split(" - ")
                    if len(parts) == 2:
                        timestamp, value = parts
                        log_data.append((timestamp, value))
                except Exception as e:
                    print(f"Skipping line due to error: {e}")
                    continue

print(log_data[:5])

# Create a DataFrame from the log data
log_df = pd.DataFrame(log_data, columns=["Timestamp", "Value"])

# Convert the Timestamp column to datetime
log_df["Timestamp"] = pd.to_datetime(log_df["Timestamp"])

print(log_df.head())

# Pivot the log DataFrame to have columns for each value type
log_df["Value Type"] = log_df["Value"].apply(
    lambda x: x.split(": ")[0] if ": " in x else None
)
log_df["Value"] = log_df["Value"].apply(
    lambda x: float(x.split(": ")[1]) if ": " in x else None
)

print(log_df.head())

# Show count of rows with NaN values
na_count = log_df.isna().sum()
print("Count of NaN values before dropping:", na_count)

# Show rows with NaN values
na_rows = log_df[log_df[["Value Type", "Value"]].isna().any(axis=1)]
print("Rows with NaN values before dropping:")
print(na_rows.head())
print(na_rows.tail())

log_df.head()

value_types_format = [
    "Count left",
    "Temp",
    "Hum",
    "PPM CO2 (MQ135)",
    "PPM CO2 (NDIR)",
    "PPM H2S (MQ136)",
    "PPM H2S (TGS2602)",
    "Analog MQ135",
    "Analog NDIR",
    "Analog MQ136",
    "Analog TGS2602",
]

additional_value_types_format = [
    "MQ135 rs_ro",
    "MQ135 rs_ro before correction",
    "MQ135 rs_ro_corr",
    "MQ135 rs_ro after correction",
    "MQ136 rs_ro",
    "MQ136 rs_ro before correction",
    "MQ136 rs_ro_corr",
    "MQ136 rs_ro after correction",
    "TGS2602 rs_ro",
    "TGS2602 rs_ro before correction",
    "TGS2602 rs_ro_corr",
    "TGS2602 rs_ro after correction",
    "NDIR Sensor",
    "NDIR Voltage",
]

print(len(value_types_format))

# Create a DataFrame with columns from value_types_format and Timestamp
log_df_pivot = pd.DataFrame(
    columns=["Timestamp"] + value_types_format + additional_value_types_format
)
print(log_df_pivot.head())

# Iterate over the rows of the log_df
for index, row in log_df.iterrows():
    if row["Value Type"] == value_types_format[0]:
        value_dict = {"Timestamp": row["Timestamp"]}
        for i_index in range(index, index + len(value_types_format)):
            if i_index < len(log_df):
                value_row = log_df.iloc[i_index]
                value_type = value_row["Value Type"]
                if value_type not in value_types_format:
                    break
                value_dict[value_type] = value_row["Value"]

        # Check for additional value types
        for i_index in range(
            index + len(value_types_format),
            index + len(value_types_format) + len(additional_value_types_format),
        ):
            if i_index < len(log_df):
                value_row = log_df.iloc[i_index]
                value_type = value_row["Value Type"]
                if value_type in additional_value_types_format:
                    value_dict[value_type] = value_row["Value"]

        log_df_pivot = pd.concat(
            [log_df_pivot, pd.DataFrame([value_dict])], ignore_index=True
        )

log_df_pivot.head()
```

```python
# print df_time columns
print(df_time.columns)

# print log_df_pivot columns
print(log_df_pivot.columns)
```

```python
# Assuming df_time and log_df_pivot are your DataFrames
log_df_merged = log_df_pivot.copy()
df_time_copy = df_time.copy()

# Ensure both timestamps have the same time zone information
df_time_copy["date created start"] = pd.to_datetime(
    df_time_copy["date created start"]
).dt.tz_convert("Asia/Bangkok")
df_time_copy["date created end"] = pd.to_datetime(
    df_time_copy["date created end"]
).dt.tz_convert("Asia/Bangkok")
log_df_merged["Timestamp"] = pd.to_datetime(log_df_merged["Timestamp"]).dt.tz_localize(
    "Asia/Bangkok"
)

# add column from df_time_copy if the timestamp log_df_merger
# is inside the range

for index, row in log_df_merged.iterrows():
    timestamp = row["Timestamp"]
    for _, time_row in df_time_copy.iterrows():
        if time_row["date created start"] <= timestamp <= time_row["date created end"]:
            log_df_merged.loc[index, "Title"] = time_row["Title"]

# Reorder columns to place "Title" after "Timestamp"
cols = list(log_df_merged.columns)
cols.insert(cols.index("Timestamp") + 1, cols.pop(cols.index("Title")))
log_df_merged = log_df_merged[cols]
```
